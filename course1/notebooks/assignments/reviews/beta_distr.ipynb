# Beta distribution
## $\alpha=0.5$
## $\beta=0.7$

```python
from scipy.stats import beta, norm
a = 0.5
b = 0.7
dist = beta(a, b)
numbers = dist.rvs(1000)
```

```python
import matplotlib.pylab as plt
import numpy as np

x = np.linspace(dist.ppf(0.01), dist.ppf(0.99), 100)
plt.hist(numbers, density=True, bins=25, alpha=0.5, label="beta distribution hist")
plt.plot(x, dist.pdf(x), "r-", lw=3, label="beta distribution pdf")
plt.ylim(0, 5)
plt.xlabel("x")
plt.ylabel("beta.pdf(x)")
plt.legend(loc="best")
plt.grid(alpha=0.2)
plt.gcf().set_size_inches(12, 8)
plt.show()
```

# Samples generation

```python
n1 = 5
n2 = 10
n3 = 50
n1_samples = np.array([dist.rvs(n1).mean() for __ in range(1000)])
n2_samples = np.array([dist.rvs(n2).mean() for __ in range(1000)])
n3_samples = np.array([dist.rvs(n3).mean() for __ in range(1000)])
```

# Graphs

```python
plt.hist(n1_samples, density=True, alpha=0.5, label="5-samples")
plt.plot(x, norm(dist.mean(), dist.std()/n1**(1/2)).pdf(x), lw=3, label="Norm distribution pdf")
plt.xlabel("x")
plt.ylabel("norm.pdf(x)")
plt.legend(loc="best")
plt.gcf().set_size_inches(12, 8)
plt.show()
```

```python
plt.hist(n2_samples, normed=True, alpha=0.5, label="10-samples")
plt.plot(x, norm(dist.mean(), dist.std()/n2**(1/2)).pdf(x), lw=3, label="Norm distribution pdf")
plt.legend(loc="best")
plt.xlabel("x")
plt.ylabel("norm.pdf(x)")
plt.gcf().set_size_inches(12, 8)
plt.show()
```

```python
plt.hist(n3_samples, normed=True, alpha=0.5, label="50-samples")
plt.plot(x, norm(dist.mean(), dist.std()/n3**(1/2)).pdf(x), lw=3, label="Norm distribution pdf")
plt.legend(loc="best")
plt.xlabel("x")
plt.ylabel("norm.pdf(x)")
plt.gcf().set_size_inches(12, 8)
plt.show()
```
