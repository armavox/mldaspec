# Центральная предельная теорема своими руками

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
import math
```

# Определяем гамма распределение

Это распределение с двумя параметрами $$(\alpha,\lambda): \alpha>0, \lambda>0$$

Мат. ожидание и дисперсия вычисляются по формулам:
$$E[X] = \alpha\lambda$$
$$D[X] = \alpha\lambda^2$$

```python
#посчитаем среднее и дисперсию
alpha = 0.7
lmbda = 3.6
E = alpha * lmbda
D = alpha * lmbda**2
print(E)
print(D)
```

Ввиду наличия у гамма распределения большого количества частных случаев, которые накладывают ограничения на входные параметры (без внесения изменений в вычисления), первый график выглядит достаточно странно (на взгляд неопытного человека(т.е. меня)).
Данная ситуация улучшается на последующих изображениях посредством ЦПТ.

```python
#генерируем выборку объема 1000
gamma = sts.gamma(0.5, loc = alpha, scale = lmbda)#задаем функцию
sample = gamma.rvs(1000)#генерим случайную выборку 1000 значений
x = np.linspace(0,10,1000)#определяем массив точек для построения плотности распределения
pdf = gamma.pdf(x)#вычисляем значения плотности распределения от выборки
#строим гистрограмму выборки и теоретическую функцию распределения
plt.plot(x, pdf, label='theoretical PDF')
plt.hist(sample, bins = 50, range=(1,10), density=True)
plt.ylabel('$f(x)$/number of samples')
plt.xlabel('$x$')
plt.legend(loc='best')
plt.show()
```

```python
n=5
sample=np.array([])#обозначим массив выборочных средних
i=1
while i <= 1000:#цикл генерации случайных выборок количества n из функции
    sample_i = gamma.rvs(n)#генерация выборки
    mean_i=sum(sample_i)/n#определение выборочного среднего (в.с.)
    sample=np.append(sample,mean_i)#добавляем в массив новое значение в.с.
    i+=1
#определим параметры нормального распределения описывающие массив 1000 в.с. при n=5
E=alpha * lmbda#среднее gamma
D=alpha * lmbda**2#дисперсия gamma
D_n=D/n#дисперсия для нормального распределения приближающего выборку средних
sigma=math.sqrt(D_n)#сигма для нормального распределения выборки средних
norm = sts.norm(E, sigma)#определяем нормальное распределение с расчетными характеристиками gamma
pdf = norm.pdf(x[:500])#получаем плотность распределения
#строим гистограмму массива средних при n=5 и плотность описывающую этот массив нормального распределения
plt.plot(x[:500], pdf, label='PDF(n=5)')
plt.hist(sample, bins =80, range=(0,5), density=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.legend(loc='best')
```
