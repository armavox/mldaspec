```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

norm = sts.norm()
x = np.linspace(-5, 5, 100)
```

```python
from matplotlib import rc
font = {'family': 'Serif',
        'weight': 'normal'}
rc('font', **font)

plt.plot(x, norm.pdf(x));
plt.xlabel('X')
plt.ylabel('Плостность распределения')
plt.title('График плотности нормального\nстандартного распределения');
```
