## Seaborn little tutorial

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['patch.force_edgecolor'] = True
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.regplot(x='total_bill', y='tip', data=tips);
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, );
```

```python
tips['big_tip'] = (tips.tip / tips.total_bill) > .15
```

```python
sns.lmplot(x='total_bill', y='big_tip',
           data=tips,
           y_jitter=.03);
plt.grid(True, which='both', alpha=.3)
```

```python
anscombe = sns.load_dataset("anscombe")
```

```python
sns.lmplot(x='x', y='y', col='dataset', col_wrap=2,
           truncate=True, data=anscombe, sharex=False, sharey=False);
```

```python
sns.lmplot(x='x', y='y', data=anscombe.query("dataset == 'I'"));
```

```python
sns.pairplot(tips, 
             hue="smoker", height=5, kind="reg");
```
