```python
import numpy as np
from scipy import stats
import seaborn as sns
sns.set_style(rc={'patch.force_edgecolor': True})
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(size=1000)
t = stats.t(100)
```

```python
fig = sns.distplot(t.rvs(1000), kde=False, fit=stats.gamma)
fig.set(xlabel='X', ylabel='Y');
```

```python
fig = sns.kdeplot(t.rvs(1000))
fig.set(xlabel='X', ylabel='Y');
```
